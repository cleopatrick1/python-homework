```python
from pathlib import Path
import csv

input_csvpath = Path('./budget_data.csv')
output_path = Path('./financial_analysis.txt')


pnl = []
dates = []
dicts_of_changes = {}
month_changed = [] 
line_num = 0
total_profit_loss = 0
counter = 0
previous_revenue = 0
total_months = 0
total_changes = 0 
average = 0
max_change = 0 
min_change = 0 
max_key = ''
min_key = ''



with open(input_csvpath, 'r') as csvfile: 
    csvreader = csv.reader(csvfile, delimiter = ',')
    next(csvreader)
    pnl = list(csvreader)
    
#MONTHS

total_months = len(pnl)


#TOTAL

for row in pnl: 
    pl = int(row[1])
    total_profit_loss = total_profit_loss + pl
    
#SUBSTRACTING PL 

    if counter > 0 : 
        change = pl - int(pnl[counter -1][1])
        dicts_of_changes[row[0]] = change
    counter = counter + 1     

#AVERAGE CHANGE
    
for key in dicts_of_changes:
    total_changes = total_changes + dicts_of_changes[key]
    if dicts_of_changes[key] >= max_change: 
        max_change = dicts_of_changes[key]
        max_key = key
    elif dicts_of_changes[key] <= min_change:
        min_change = dicts_of_changes[key]
        min_key = key
    
    
average = round(total_changes / len(dicts_of_changes),2)


#GREATEST INCREASE


result = "Financial Analysis \n---------------------------------\n\n"
result += f"Total Months: {total_months}\n"
result += f"Total: ${total_profit_loss}\n"
result += f"Average Change: ${average}\n"
result += f"Greatest Increase in Profits: {max_key} (${max_change})\n"
result += f"Greatest Decrease in Profits: {min_key} (${min_change})"
    
print(result)
    
    
with open(output_path , 'w') as csvfile: 
    csvfile.write(result)
    
    
    
```
